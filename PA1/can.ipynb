```python
import os
os.chdir('/content/drive/MyDrive/comp5214/PA1')
from data_loader import *
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import tensorflow as tf

(train_images, train_labels), (test_images, test_labels) = load_mnist()

# Define the number of channels to test
num_of_channels = [8, 16, 32, 64, 128]
accuracies = []
batch_size = 64
num_epoch = 20

def can(num_of_channel):
    # Define the model
    model = models.Sequential([
        layers.Conv2D(num_of_channel, (3, 3), activation='leaky_relu', padding='same', input_shape=(28, 28, 1)),
        layers.Conv2D(num_of_channel, (3, 3), dilation_rate=2, activation='leaky_relu', padding='same'),
        layers.Conv2D(num_of_channel, (3, 3), dilation_rate=4, activation='leaky_relu', padding='same'),
        layers.Conv2D(num_of_channel, (3, 3), dilation_rate=8, activation='leaky_relu', padding='same'),
        layers.Conv2D(10, (3, 3), activation='leaky_relu', padding='same'),
        layers.GlobalAveragePooling2D(),
        layers.Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    model.fit(train_images, train_labels, epochs=num_epoch, batch_size=batch_size, verbose=0)

    test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
    accuracies.append(test_acc)
    print(f"Accuracy with {num_of_channel} channels: {test_acc:.4f}")

for n in num_of_channels:
    can(n)

plt.figure()
plt.plot(num_of_channels, accuracies, marker='o')
plt.title('Accuracy vs Number of channels')
plt.xlabel('Number of channels')
plt.ylabel('Accuracy')
plt.xticks(num_of_channels)
plt.grid()
plt.show()
```
